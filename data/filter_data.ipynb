```python
import json
import pandas as pd 
from pandas.io.json import json_normalize 
import csv
```

```python
with open('kobe_data.json') as json_file:
    data = json.load(json_file)
```

```python
pd.set_option('display.max_columns', None)

kobe_data = pd.json_normalize(data)
kobe_data.head(10)
```

```python
kobe_data.to_csv(r'kobe_data.csv', index=False)
```

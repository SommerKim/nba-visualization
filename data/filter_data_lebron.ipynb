```python
import json
import pandas as pd 
from pandas.io.json import json_normalize 
import csv
```

```python
with open('lebron_data.json') as json_file:
    data = json.load(json_file)
```

```python
pd.set_option('display.max_columns', None)

lebron_data = pd.json_normalize(data)
lebron_data.head(10)
```

```python
lebron_data.to_csv(r'lebron_data.csv', index=False)
```
